```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
file = open("output/optimized_gene_parameters.txt", "r")
data = []
for line in file:
    data.append(line.strip().split("\t"))
    
data_np = np.array(data)
# print(data_np)
data_np_clean = data_np[data_np[:,1] != "nan"]
data_np_clean = data_np_clean[data_np_clean[:,1] != "NA"]
data_np_clean = data_np_clean.astype("float") 
```

```python
# load the data on the gene expression
data = np.load("processed_data/norm_filtered_cells.scaled.pickle", allow_pickle=True)
num_cells = data.shape[2]
num_genes = data.shape[1]
```

```python
# Load the gene names
genes = np.load("processed_data/gene_name.filtered.pickle", allow_pickle=True)
```

```python
# Sort data by gamma values
data_np_clean_sorted_desc = data_np_clean[np.argsort(data_np_clean[:,2])[::-1]]
```

```python
num_genes = 10
plt.bar(range(10), data_np_clean_sorted_desc[0:num_genes,2], 
        tick_label=genes[data_np_clean_sorted_desc[0:num_genes,0].astype("int")],
       color="orange")
plt.xlabel("Genes")
plt.ylabel("Genes with the highest rate \n of degradation (gamma)")
plt.title("Top 10 genes with the highest degradation rate")
plt.xticks(rotation='vertical')
plt.tight_layout()
plt.savefig("figure/degradation_top10_highest.png", bbox_inches = 'tight')
```

```python
num_genes = 10
plt.bar(range(10), data_np_clean_sorted_desc[-num_genes:,2],
        tick_label=genes[data_np_clean_sorted_desc[-num_genes:,0].astype("int")],
       color="orange")
plt.xlabel("Genes")
plt.ylabel("Genes with the lowest rate \n of degradation (gamma)")
plt.title("Top 10 genes with the lowest degradation rate")
plt.xticks(rotation='vertical')
plt.tight_layout()
plt.savefig("figure/degradation_top10_lowest.png", bbox_inches = 'tight')
```

```python
cell_order = np.loadtxt("output/cell_order.txt", delimiter=",")
cell_order = cell_order.astype("int") 

def running_mean(x, N=100):
    cumsum = np.cumsum(np.insert(x, 0, 0)) 
    return (cumsum[N:] - cumsum[:-N]) / float(N)
    
# np.array(smooth)

# cell_order.shape
# cell_order
# len(set(cell_order))
```

```python
gene_idx = 200
gene_name = genes[gene_idx]


smooth_spliced = running_mean(data[0,gene_idx,cell_order])
smooth_unspliced = running_mean(data[1,gene_idx,cell_order])
plt.plot(range(len(smooth_unspliced)), smooth_unspliced, label="unspliced (reactant)")
plt.plot(range(len(smooth_spliced)), smooth_spliced, label="spliced (product)")
plt.xlabel("Pseudo-time (t)")
plt.ylabel("Gene expression")
plt.legend()
plt.title("Gene expression with time for %s" %gene_name)
plt.savefig("figure/pseudotime_example1.png")
```
